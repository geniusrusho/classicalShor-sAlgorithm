```python
import math
import matplotlib.pyplot as mpl
import numpy as np
```

```python
N = 15
a = 13

#N = 21
#a = 2
```

```python
math.gcd(a, N)
```

```python
z = list(range(N))
y = [a**z0 % N for z0 in z]

mpl.plot(z,y)
mpl.xlabel('z')
mpl.ylabel(f'{a}^z (mod {N})')
mpl.show()
```

```python
r = z[y[1:].index(1)+1]
print(r)
```
